## Section II: Homogenous periodic simulations 

## Section II.A:  Weak coupling in homogenous periodic simulations

The paper specifies these parameters:

* Periodic box.
* Laser frequency $\omega_0/\omega_{pe} = 2.5$.
* $v_0/c = 0.03$.
* $v_e/c = 0.0357$.
* Fixed ions.
* Box length $L = 5.464 c/\omega_{pe}$.
* Total simulated time $T = 600 \omega_{pe}^{-1}$.
* The simulation uses 20,000 particles.

Further comments:

* It is easier for our diagnostic purposes to use a linearly polarized laser.
* For this particular simulation I have made the laser incident from the left (from x=0 side)
* The above gives $n_e/n_{cr} = 0.16$.
* For $\omega_{pe} = 1$, we have $\lambda_D = 0.0357$, so we use:
  * 154 cells, with cell size $\Delta x = 5.464/154 = 0.0355$. 
  * Time step $\Delta t = 0.035$.
  * 130 particles per cell, or $130 * 154 = 20020$ total particles.
* For backward SRS, $k\lambda_D = 0.12$.
* For $\lambda_0 = 351$ nm, the laser has an intensity of approximately $1 \times 10^{16}$ W/cm$^2$.
