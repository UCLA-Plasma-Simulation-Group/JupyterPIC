# Electron Trajectories due to a relativistic Particle Beam

## Theory

### Determining Electromagnetic fields from a relativistic particle beam

In this notebook, we will derive the equations of motion for an electron due to a relativistic particle beam ($\gamma_b \gg 1$, $\vec{v_b} =  v_b \hat{z} \simeq c \hat{z}$). We will first start by describing the fields $(\phi, \vec{A})$ due to the space charge of the particle beam $(\rho_b, \vec{J}_b)$. The goal is to derive constants of motion from the force equations and expressions for the final momenta of electrons after passing through the beam driver. For these calculations, we will use the Lorentz Gauge:

$$\begin{equation} \left(\frac{1}{c^2} \frac{\partial^2 }{\partial t^2}- {\nabla^2}\right) \begin{pmatrix}\phi \\ \vec{A} \end{pmatrix} = 4\pi \begin{pmatrix} \rho_b \\ \frac{\vec{J_b}}{c} \end{pmatrix} \tag{1} \end{equation}$$

$$\nabla \cdot \vec{A} + \frac{1}{c} \frac{\partial \phi}{\partial t} = 0 \tag{2}$$

We also use the quasi-static approximation to describe the potentials and current/charge of the particle beam: 

$$\begin{pmatrix} \phi \\ \vec{A} \\ \rho_b \\ \vec{J_b} \end{pmatrix} (x,y,\xi \equiv ct-z,s \equiv z) \tag{3} $$

While we have chosen the coordinate $\xi = ct-z$ to describe our relevant driver quantities, we leave it as an exercise for students to show that choosing an exact coordinate $\zeta = v_b t -z$ would produce a small corrective term in Eq. (1) which can be neglected $\left(\frac{1}{\gamma_b^2} \frac{\partial}{\partial \zeta^2}\ll \nabla_\perp^2 \right)$.   Expanding the LHS of Eq. (1), we find

$$ \left(\frac{1}{c^2} \frac{\partial^2 }{\partial t^2}- {\nabla^2}\right) = \left( \frac{1}{c^2} \frac{\partial^2 }{\partial t^2} - \frac{\partial^2 }{\partial z^2}- \nabla_\perp^2 \right)=  - \nabla_\perp^2 \tag{4}$$

Substituting the current $\vec{J} = c\rho_b \hat{z}$ into Eq. (1), we can simplify further

$$ \begin{align} 
 - \nabla_\perp^2 \phi &= 4 \pi \rho_b  \tag{5} \\
 - \nabla_\perp^2 \vec{A} &= 4 \pi \rho_b \hat{z} \tag{6}
\end{align} $$

It is worth noting that solutions to these poisson-like equations can be found by integrating over gaussian surfaces.  Subtracting Eqs. (5) & (6), we get

$$ -\nabla_\perp^2 \left(\phi-A_z\right) = 0 \tag{7}$$
$$ \boxed{\phi = A_z} \tag{8}$$

We can also simplify Eq. (2) using our quasi-static formalism, 

$$ \begin{align}
\nabla \cdot \vec{A} + \frac{1}{c} \frac{\partial \phi}{\partial t} = \nabla_\perp A_\perp + \frac{\partial A_z}{\partial z} + \frac{1}{c} \frac{\partial \phi}{\partial t} \tag{9} \\ 
 0 = \nabla_\perp A_\perp + \frac{\partial}{\partial \xi}\left(\phi - A_z\right) \tag{10} \\ 
\end{align}$$

Plugging Eq. (8) into Eq. (10), we find 

$$ \boxed{ A_x = A_y = 0} \tag{11}$$

Using Eqs. (8) and (11), we can now write expressions for the electric and magnetic fields

$$ \boxed{E_z = -\frac{\partial \phi}{\partial z} - \frac{1}{c} \frac{\partial A_z}{\partial t} = \frac{\partial}{\partial \xi} \left(\phi - A_z\right) \cong 0} \tag{12} $$

$$ \boxed{E_r(r,\xi) = B_\phi(r,\xi) = -\frac{\partial \phi}{\partial r} = \frac{4 \pi}{r} \int_0^r \rho_b(r',\xi) \ r' dr' } \tag{13} $$

### Constants of Motion

We now write the equations of motion for a charged particle $q$,  

$$\frac{d p_r}{dt} = q\left(E_r - \frac{v_z}{c} B_{\phi}\right) \tag{14} = qE_r\left(1-\frac{v_z}{c}\right) $$
$$\frac{d p_z}{dt} = q\frac{v_r}{c}B_{\phi} \tag{15} = q\frac{v_r}{c}E_r $$
$$\frac{d}{dt} \gamma mc^2 = \vec{F} \cdot \vec{v} = qv_r E_r \tag{16} $$

Moving forward, we will use the notation $m=c=1$ for simplicity. Subtracting Eqs. (15) and (16), we find the first constant of motion

$$ \frac{d}{dt}(\gamma - p_z) = 0 \tag{17} $$

Therefore, for an electron initially at rest $(\gamma_0 =1)$ we find a relation between $p_z$ and $\gamma$

$$ \boxed{ \gamma -p_z = 1 }  \tag{18}$$

Squaring both sides of Eq. (18), we can also solve for $p_z$ in terms of $p_r$

$$\begin{align}
 \gamma^2 &= (1+p_z)^2 \tag{19} \\
  1+p_r^2 + p_z^2 &= 1 + 2p_z + p_z^2 \tag{20} \\ 
\end{align} $$ 
$$  \boxed{p_z = \frac{1}{2}p_r^2}  \tag{21}$$

Next, we are interested in determining how $p_r$ varies as a function of the driver potential $\phi(r,\xi)$. We start by multiplying both sides of Eq. (14) by $p_r$
$$ p_r \left[\frac{d}{dt} p_r = qE_r(1-v_z) \right]  \tag{22}$$
$$\begin{align} 
    \frac{1}{2}\frac{d}{dt} p_r^2 &= -q\frac{\partial \phi}{\partial r}\gamma(1-v_z)v_r  \tag{23} \\
\end{align}$$

Subsituting in Eq. (18), we find

$$ \frac{1}{2}\frac{d}{dt} p_r^2 = -q\frac{\partial \phi}{\partial r}\frac{dr}{dt} $$

To simplify this expression, we can write the total time derivative of $\phi$ as 

$$ \frac{d }{dt}\phi(r,\xi) = \frac{\partial \phi}{\partial r}\frac{dr}{dt} + \frac{\partial \phi}{\partial \xi}\frac{d\xi}{dt}  \tag{24}$$

Plugging Eq. (24) into (23), we get

$$\begin{align}
\frac{d}{dt} \frac{1}{2} p_r^2 &= -q\left(\frac{d\phi}{dt} - \frac{\partial \phi}{\partial \xi}\frac{d\xi}{dt}\right) \tag{25} \\ 
\frac{d}{dt} \left[\frac{1}{2} p_r^2 + q\phi \right] &= q\frac{\partial \phi}{\partial \xi}\frac{d\xi}{dt} \tag{26} \\
\end{align} $$


Mutiplying both sides by $\frac{dt}{d\xi}$, we finally obtain

$$ \boxed{ \frac{d}{d\xi} \left[\frac{1}{2} p_r^2 + q\phi \right] = q\frac{\partial \phi}{\partial \xi} } \tag{27} $$

For an electron initially at rest $(\vec{p_i} = 0)$, we can integrate Eq. (27)

$$ \int_{- \infty}^{\xi} d\left[\frac{1}{2} p_r^2  - e\phi \right] = \int_{- \infty}^{\xi}-e\frac{\partial \phi}{\partial \xi'}d\xi' \tag{28} $$

$$ \left[ \frac{1}{2} p_r^2  - e \phi \right] = -e\int_{- \infty}^{\xi} \frac{\partial \phi}{\partial \xi'}d\xi' \tag{29}$$ 

### Relativistic regime for beam-driven PWFA

We are interested in solving Eq. (29) for a simple finite flat-top density profile $n_b(r,\xi) = n_{b0} \left(H(\xi - \xi_i) - H(\xi-\xi_f) \right) H(a_0 - r)$, where $H(x)$ represents the heaviside function. This profile describes a constant charge density $n_{b0}$ within $\xi_i \leq \xi \leq \xi_f$ and $r \leq a_0$.  We leave it as an exercise for to show that $\frac{\partial \phi}{\partial \xi} = \left[ \delta(\xi-\xi_i) - \delta(\xi-\xi_f) \right] \phi(r)$. We can now obtain an expression for $p_r$ after the electron has passed through the particle beam $(\xi > \xi_f)$, 

$$p_r^2 = 2e\left(\phi(r_f) - \phi(r_i) \right) \tag{30}$$

$$\boxed{p_r = \sqrt{2e\Delta \phi}} \tag{31}$$

We consider an electron which is initially at $r > a_0$. We leave it as an exercise to show that by integrating Eq. (13) the Electric field in the region $r > a_0$ is $E_r = -4 \pi e n_{b0} a_0^2/2r$. The potential difference is therefore given by 

$$ \begin{align}
\frac{e\Delta \phi}{mc^2} &= -\frac{e}{mc^2}\int_{r_i}^{r_f} dr E_r \tag{32}\\ 
\frac{e\Delta \phi}{mc^2} &= \frac{2\pi e^2 n_{b0} a_0^2}{mc^2} \ln \left(\frac{r_f}{r_i}\right) \tag{33}\\ 
\end{align}$$

Finally, plugging back into Eq. (31),

$$\boxed{ \frac{p_r}{mc} = \sqrt{\frac{4 \pi e^2 n_{b0} a_0^2}{mc^2} \ln \left(1 + \frac{\Delta r}{r_i}\right) } }  \tag{34}$$

where $\Delta r = r_f - r_i$ and will always be less than the beam length $\Delta r < \xi_f - \xi_i$.  It turns out that the quantity under the square root corresponds to the normalized charge per unit length $\Lambda$

$$ \Lambda(\xi) \equiv \frac{4 \pi e^2}{mc^2} \int_0^{\infty} n_b(r',\xi) \ r' dr' = \frac{ 2 \pi e^2 n_{b0} a_0^2}{mc^2} \tag{35}$$

$$\boxed{ \frac{p_r(\xi \geq \xi_f)}{mc} = \sqrt{ 2 \Lambda \ln \left(1 + \frac{\Delta r}{r_i}\right)} } \tag{36}$$

Therefore, assuming $\ln \left(1 + \frac{\Delta r}{r_i}\right) \sim O(1)$, relativistic momenta can be achieved when $\Lambda > 1$.


### Electron Trajectory Simulations (OSIRIS)

We have included some example with a flat top driver below. OSIRIS simulations are done in normalized units. 

* Time:  $t' = t \omega_{p}$

* Frequency: $\omega' = \frac{\omega}{\omega_{p}}$ 

* Position: $\vec{x}' = \frac{\omega_{p}}{c} \vec{x}$  

* Momenta: $\vec{u}' = \frac{\vec{p}}{m_{e} c} = \frac{\vec{\gamma v}}{c} = \frac{\vec{u}} {c}$  

* Electric field: $\vec{E}' = e \frac{c / \omega_{p}}{m_{e} c^2} \vec{E}$  

* Magnetic field: $\vec{B}' = e \frac{c / \omega_{p}}{m_{e} c^2} \vec{B}$
